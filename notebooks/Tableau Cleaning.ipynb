# Raw Data 

## Amsterdam

```python
amsterdam_weekdays = pd.read_csv("../data/raw/amsterdam_weekdays.csv")
amsterdam_weekdays["weekend"] = False
amsterdam_weekdays["city"] = "Amsterdam"
amsterdam_weekdays
```

```python
amsterdam_weekends = pd.read_csv("../data/raw/amsterdam_weekends.csv")
amsterdam_weekends["weekend"] = True
amsterdam_weekends["city"] = "Amsterdam"
amsterdam_weekends
```

```python
amsterdam = pd.concat([amsterdam_weekdays, amsterdam_weekends], ignore_index = True)
#amsterdam.to_csv("../data/raw/amsterdam.csv", index=False) 
amsterdam
```
